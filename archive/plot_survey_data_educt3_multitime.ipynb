## Purpose

This Jupyter Notebook uses a previously imported CSV file containing data from the Household Pulse Survey, "designed to quickly and efficiently deploy data collected on how people’s lives have been impacted by the COVID-19 pandemic." Visit this website for more detail: https://www.census.gov/programs-surveys/household-pulse-survey/data.html  

Here, we create simple plots to explore the data. 


## Set up environment

```python
# Import modules 
import pandas as pd # data manipulation
import plotly.express as px # data viz with plotly 
```

```python
# Import data
income_df = pd.read_csv('educ3_income_multitime.csv')
food_df = pd.read_csv('educ3_food_multitime.csv')

# Visually inspect data
income_df
```

## Plot data

```python
food_df["% Internet always available for educational purposes"] = food_df["Internet always available for educational purposes"]/food_df["Total"]

fig = px.line(food_df[food_df["Location"] == "WA"], x = "Date", y = "% Internet always available for educational purposes", color = "Group")
fig.show()
```

```python
food_df["% Internet always available for educational purposes"] = food_df["Internet always available for educational purposes"]/food_df["Total"]

fig = px.line(food_df[food_df["Location"] == "Seattle_Metro_Area"], x = "Date", y = "% Internet always available for educational purposes", color = "Group")
fig.show()
```

```python
food_df["% Device always available for educational purposes"] = food_df["Device always available for educational purposes"]/food_df["Total"]

fig = px.line(food_df[food_df["Location"] == "WA"], x = "Date", y = "% Device always available for educational purposes", color = "Group")
fig.show()
```

```python
food_df["% Device always available for educational purposes"] = food_df["Device always available for educational purposes"]/food_df["Total"]

fig = px.line(food_df[food_df["Location"] == "Seattle_Metro_Area"], x = "Date", y = "% Device always available for educational purposes", color = "Group")
fig.show()
```
