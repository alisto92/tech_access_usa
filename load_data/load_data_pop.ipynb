# Tech Access in the USA 

## Purpose

The purpose of this notebook is to read in historical data on the percent of population living in rural verus urban populations by state in the United States. The data were originally downloaded from this website: https://www.icip.iastate.edu/tables/population/urban-pct-states 

Joins to file with state abbreviations originally found here: https://gist.github.com/dantonnoriega/bf1acd2290e15b91e6710b6fd3be0a53 

## Set up environment

```python
# Import modules 
import pandas as pd # data manipulation
```

```python
# Initialize list of file names 
# Raw files
raw_dir = '/home/jupyter/tech_access_usa/raw_data/percent_urban_pop/' # directory
# Save files
save_dir = '/home/jupyter/tech_access_usa/csv_data/' # directory
save_file = 'urban_pop.csv'
```

## Import data

```python
# Import population
urban_pop = pd.read_excel(raw_dir + "pop-urban-pct-historical.xls", sheet_name = "States", 
                          skiprows = 5, skipfooter = 7)
# Visual check 
urban_pop.tail()
```

```python
# Import abbreviation data  
abbv = pd.read_csv(raw_dir + "us-state-ansi-fips.csv")
abbv.tail()
```

## Merge data

```python
# Merge population & abbreviation dataframe, get rid of extra columns & rename column
urban_pop_df = pd.merge(urban_pop[['FIPS', 'Area Name', 2010]], abbv, how = 'left', left_on =['FIPS', 'Area Name'], right_on = [' st', 'stname']).drop([' st', 'stname'], axis=1).rename(columns={'Area Name':"area_name", 2010:"pct_urban_2010", ' stusps':"area_abbv"})
```

```python
# Create abbreviation for United States that matches U.S. Census data 
urban_pop_df.iloc[0,3] = "US"
```

```python
# Visual inspection 
urban_pop_df
```

## Visualization

```python
# Create histogram of percent urban in 2010 
urban_pop_df.hist(column='pct_urban_2010', bins = 30)
```

## Save data

```python
# Save to CSV 
urban_pop_df.to_csv(save_dir + save_file, index=False)
```
